```python
import sklearn.datasets as datasets
import sklearn.model_selection as model_selection
import sklearn.tree as trees
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pandas as pd
```

```python
def showheader():
    data = datasets.load_iris()
    df = pd.DataFrame(data.data, columns=data.feature_names)
    print(df.head())

showheader()    
```

```python
def groupbarchart():
    labels = ['sepal length', 'sepal width', 'petal length', 'petal width']
    data = datasets.load_iris()
    df = pd.DataFrame(data.data, columns=data.feature_names)
    df['target'] = data.target
    df['target'] = df['target'].apply(lambda x: data.target_names[x])
    df = df.groupby('target').mean()
    df.plot(kind='bar')
    plt.show()

groupbarchart()    
```

```python
mpl.rcParams['figure.figsize'] = [15, 10]


def tree():
    data = datasets.load_iris()
    Xtrain, Xtest, ytrain, ytest = model_selection.train_test_split(data.data, data.target, test_size=0.2)
    model = trees.DecisionTreeClassifier()
    model.fit(Xtrain, ytrain)
    trees.plot_tree(model)
    plt.show()
    
tree()
```

```python
def radarplot():
    labels=np.array(['A', 'B', 'C', 'D'])
    stats=np.array([60,70,80,90])
    stats2=np.array([90,80,70,60])
    angles=np.linspace(0, 2*np.pi, len(labels), endpoint=False).tolist()
    stats=np.concatenate((stats,[stats[0]]))
    stats2=np.concatenate((stats2,[stats2[0]]))
    angles+=angles[:1]
    _, ax = plt.subplots()
    ax = plt.subplot(111, polar=True)
    ax.fill(angles, stats, 'b', alpha=0.1)
    ax.fill(angles, stats2, 'r', alpha=0.1)
    ax.set_yticklabels([])
    plt.show()


radarplot()    
```

```python
def boxplot():
    data = datasets.load_diabetes()
    Xtrain, Xtest, ytrain, ytest = model_selection.train_test_split(data.data, data.target, test_size=0.2)
    plt.boxplot(Xtrain)
    plt.show()

boxplot()    
```

```python
def scatterplot():
    data = datasets.load_iris()
    Xtrain, Xtest, ytrain, ytest = model_selection.train_test_split(data.data, data.target, test_size=0.2)
    plt.scatter(Xtrain[:,0], Xtrain[:,1], c=ytrain)
    plt.show()

scatterplot()    
```
